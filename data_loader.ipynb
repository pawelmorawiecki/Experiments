```python
import nibabel as nib
import numpy as np
import h5py
import os
from Utils import save_large_dataset, load_large_dataset
```

```python
indir = 'data/'
Xs = []
for root, dirs, filenames in os.walk(indir):
    filenames.sort()
    for f in filenames:
        if '.nii' == f[-4:]:
            img = nib.load(indir + f)
            data = img.dataobj # get the data object
            X = np.expand_dims(data, -1)
            X = X / np.max(X) # normalize to (0,1) range
            X = X.astype('float32')
            X = np.expand_dims(X, 0) #add one extra dimension for channel
            Xs.append(X)

Xa = np.vstack(Xs)
```
